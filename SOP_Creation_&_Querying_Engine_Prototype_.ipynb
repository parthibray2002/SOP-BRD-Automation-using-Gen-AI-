```python
import os
from google.colab import drive
from langchain_community.document_loaders import PyPDFLoader, Docx2txtLoader

# 1. Mount your Google Drive (This will prompt you to log in and authorize)
drive.mount('/content/drive')

def load_documents(directory_path):
    documents = []
    # Check if the path exists before trying to read it
    if not os.path.exists(directory_path):
        print(f"Error: The path {directory_path} does not exist. Check your folder name.")
        return documents

    for filename in os.listdir(directory_path):
        filepath = os.path.join(directory_path, filename)
        if filename.endswith(".pdf"):
            loader = PyPDFLoader(filepath)
            documents.extend(loader.load())
        elif filename.endswith(".docx"):
            loader = Docx2txtLoader(filepath)
            documents.extend(loader.load())
    return documents

# 2. Update this path to point to the specific folder in your Google Drive
# 'MyDrive' is the root of your Google Drive.
drive_folder_path = "/content/drive/MyDrive/Automation Files"

# Load all historical project data and manuals
raw_docs = load_documents(drive_folder_path)
print(f"Loaded {len(raw_docs)} document pages.")
```

# 3. SOP Generation
Using GPT-4 to synthesize a standardized SOP based on the uploaded source documents.

```python
!pip install -q langchain-openai python-docx

from langchain_openai import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from docx import Document

def generate_sop(docs):
    # Combine the extracted text to feed to the LLM (ensure it fits context window)
    context_text = "\n".join([doc.page_content for doc in docs[:15]])

    llm = ChatOpenAI(temperature=0.2, model_name="gpt-4-turbo")

    sop_prompt = PromptTemplate.from_template("""
    You are an expert Project Manager. Based on the following source documents, generate a comprehensive Standard Operating Procedure (SOP) for Site Setup.

    It MUST include the following 15 sections:
    1. Scope & Objective, 2. Site Organisation, 3. Project Contract Details, 4. Human Resources, 5. Site Admin/Facilities, 6. Planning & Monitoring, 7. Documentation/Filing, 8. Interfaces, 9. Quality Control, 10. Stores, 11. Commercial, 12. EHS, 13. Statutory Compliance, 14. Site Closure, 15. Annexures.

    Also include a "Document Control" header and a "Lessons Learnt" section.

    Source Context:
    {context}

    Draft the SOP:
    """)

    print("Generating SOP...")
    chain = sop_prompt | llm
    sop_content = chain.invoke({"context": context_text}).content
    return sop_content

def save_sop_to_word(content, filename="Automated_Site_Setup_SOP.docx"):
    doc = Document()
    doc.add_heading('Standard Operating Procedure: Site Setup', 0)
    doc.add_paragraph(content)
    doc.save(filename)
    print(f"SOP successfully saved to {filename}")

# Execute SOP Generation
sop_text = generate_sop(raw_docs)
save_sop_to_word(sop_text)
```

# 4. Knowledge Base Construction
Chunking documents and creating a searchable vector database using LlamaIndex.

```python
!pip install -q faiss-cpu

from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS

def build_vector_database(raw_docs, sop_text):
    # 1. Add the newly generated SOP to our knowledge base
    from langchain_core.documents import Document as LangchainDocument
    sop_doc = LangchainDocument(page_content=sop_text, metadata={"source": "Automated_SOP", "type": "Policy"})
    all_docs = raw_docs + [sop_doc]

    # 2. Split documents into smaller semantic chunks
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=150,
        separators=["\n\n", "\n", " ", ""]
    )
    chunks = text_splitter.split_documents(all_docs)

    # 3. Create Embeddings and Store in FAISS Vector DB
    print(f"Embedding {len(chunks)} chunks into Vector DB...")
    embeddings = OpenAIEmbeddings()
    vector_db = FAISS.from_documents(chunks, embeddings)

    # Save the DB locally so we don't have to rebuild it every time
    vector_db.save_local("faiss_knowledge_base")
    print("Vector Database built and saved successfully.")
    return vector_db

# Execute Vector DB Creation
vector_db = build_vector_database(raw_docs, sop_text)
```

```python
# If you haven't already: !pip install -q llama-index

import os
from llama_index.core import Document as LlamaDocument, VectorStoreIndex, Settings
from llama_index.core.node_parser import SentenceSplitter

def build_vector_database(raw_docs, sop_text):
    # 1. Convert any LangChain documents to LlamaIndex documents
    llama_docs = []
    for d in raw_docs:
        # Check if it's a LangChain doc (has page_content attribute)
        if hasattr(d, 'page_content'):
            llama_docs.append(LlamaDocument(text=d.page_content, metadata=d.metadata))
        else:
            llama_docs.append(d)

    # 2. Add the newly generated SOP to our knowledge base
    sop_doc = LlamaDocument(
        text=sop_text,
        metadata={"source": "Automated_SOP", "type": "Policy"}
    )
    all_docs = llama_docs + [sop_doc]

    # 3. Configure chunking settings (LlamaIndex v0.10+ style)
    Settings.text_splitter = SentenceSplitter(
        chunk_size=1000,
        chunk_overlap=150
    )

    # 4. Create Embeddings and Store in Vector DB
    print("Processing, chunking, and embedding documents into LlamaIndex...")
    index = VectorStoreIndex.from_documents(all_docs)

    # 5. Save the DB locally
    persist_dir = "./llama_knowledge_base"
    index.storage_context.persist(persist_dir=persist_dir)
    print(f"Vector Database built and saved successfully to '{persist_dir}'.")

    return index

# Execute Vector DB Creation
index = build_vector_database(raw_docs, sop_text)
```

# 5. AI Query Engine & Escalation Logic
Setting up the RAG pipeline with built-in confidence checks for human escalation.

```python
from llama_index.core import StorageContext, load_index_from_storage

def query_sop_system(query_text, index):
    # Create a query engine
    query_engine = index.as_query_engine(similarity_top_k=3)

    # Modify the prompt to include the LOW_CONFIDENCE instruction
    custom_prompt = (
        f"Use the provided context to answer the question: {query_text}. "
        "If the answer is not in the context, output exactly: 'LOW_CONFIDENCE'. "
        "Always mention the source document from the metadata if possible."
    )

    response = query_engine.query(custom_prompt)
    response_text = str(response)

    if "LOW_CONFIDENCE" in response_text.upper():
        print("\n--- ESCALATION TRIGGERED ---")
        return "I couldn't find a confident answer in the current documents. Escalating to a human manager."

    return response_text

# 1. Test Site Closure Query
query_1 = "What are the key steps for Site Closure according to the SOP?"
print(f"Q: {query_1}")
print(f"A: {query_sop_system(query_1, index)}\n")

# 2. Test Escalation (Out of context query)
query_2 = "How do I bake a chocolate cake?"
print(f"Q: {query_2}")
print(f"A: {query_sop_system(query_2, index)}")
```

# 6. Interactive Chatbot Interface
Launching a visual UI to interact with the automated SOP assistant.

```python
!pip install -q gradio

import gradio as gr

def chatbot_interface(question, history):
    # Re-using the logic from our query_sop_system
    try:
        response = query_sop_system(question, index)
        return response
    except Exception as e:
        return f"Error: {str(e)}"

# Create the Gradio UI
view = gr.ChatInterface(
    fn=chatbot_interface,
    title="SOP Automation Assistant",
    description="Ask any questions regarding Site Setup, SOPs, or Project Management protocols.",
    examples=["What are the key steps for Site Closure?", "What is required for Site Admin?"],
    theme="soft"
)

view.launch(share=True, debug=True)
```

# 7. Deployment as Streamlit Web App
Creating a standalone web application for the SOP Assistant.

```python
%%writefile app.py
import streamlit as st
import os
from llama_index.core import StorageContext, load_index_from_storage

st.set_page_config(page_title="SOP AI Assistant", page_icon="🏗️", layout="centered")

# Custom CSS for a professional Chatbot feel
st.markdown("""
    <style>
    .main {
        background-color: #f5f7f9;
    }
    .stChatMessage {
        border-radius: 15px;
        padding: 10px;
        margin-bottom: 10px;
    }
    .stChatInputContainer {
        padding-bottom: 20px;
    }
    h1 {
        color: #1E3A8A;
        font-family: 'Segoe UI', Tahoma, Geneva, Verdana, sans-serif;
    }
    </style>
    """, unsafe_allow_html=True)

# Helper to load index
@st.cache_resource
def load_kb_index():
    storage_context = StorageContext.from_defaults(persist_dir="./llama_knowledge_base")
    return load_index_from_storage(storage_context)

# Sidebar Enhancements
with st.sidebar:
    st.title("🛠️ Control Center")
    st.info("This AI Assistant is trained on internal Site Setup SOPs and historical project data.")

    try:
        index = load_kb_index()
        st.success("✅ Knowledge Base Active")
    except:
        st.error("❌ Knowledge Base Offline")

    st.divider()
    st.subheader("Project Details")
    st.write("**Current SOP Version:** 1.0")
    st.write("**Last Updated:** Feb 2026")

    if st.button("Clear Chat History"):
        st.session_state.messages = []
        st.rerun()

# Main UI
st.title("🏗️ SOP AI Assistant")
st.caption("Enterprise-grade Project Management Intelligence")

if "messages" not in st.session_state:
    st.session_state.messages = [{"role": "assistant", "content": "Hello! I am your Project SOP Assistant. How can I help you with site setup or closure today?"}]

# Display messages
for message in st.session_state.messages:
    with st.chat_message(message["role"]):
        st.markdown(message["content"])

# Chat Logic
if prompt := st.chat_input("Type your question here..."):
    st.session_state.messages.append({"role": "user", "content": prompt})
    with st.chat_message("user"):
        st.markdown(prompt)

    with st.chat_message("assistant"):
        with st.spinner("Consulting internal documents..."):
            query_engine = index.as_query_engine(similarity_top_k=3)
            custom_prompt = (
                f"Answer the following based on internal SOPs: {prompt}. "
                "If unknown, strictly output 'LOW_CONFIDENCE'."
            )
            response = str(query_engine.query(custom_prompt))

            if "LOW_CONFIDENCE" in response.upper():
                response = "⚠️ **Human Escalation Triggered**\n\nI couldn't find a definitive answer in our current Standard Operating Procedures. This query has been flagged for review by the Project Management Office (PMO)."

            st.markdown(response)
            st.session_state.messages.append({"role": "assistant", "content": response})
```
